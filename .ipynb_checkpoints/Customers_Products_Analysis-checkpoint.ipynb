### 2.3 Identify missing or inconsistent data
Missing values and inconsistent data are often addressed during the data collection phase, as databases are typically designed to enforce constraints that prevent missing or invalid data. If a column allows missing values or duplicates, it is usually because the database was intentionally designed to accommodate such cases. However, in this project, we will demonstrate how to identify missing values in SQL. As for inconsistent data, such as inconsistent column names or formats, corrections are typically made during querying by using techniques like aliasing or data transformation. We will use **payments** table to demonstrate how missing values are identified in SQL.

```python
# Query for checking missing values
query = """
SELECT COUNT(*)-COUNT(customerNumber) AS customerNumber,
       COUNT(*)-COUNT(checkNumber) AS checkNumber,
       COUNT(*)-COUNT(paymentDate) AS paymentDate,
       COUNT(*)-COUNT(amount) AS amount
  FROM payments;
        """
check_missing_values = pd.read_sql(query, conn).T.rename(columns = {0:'Missing Values'})
check_missing_values
```

From the output above, we can easily see that our table **payments** has no missing values. 

### 2.4 Identify duplicate rows
The following SQL query demonstrates how to check for duplicate records in a database. We will use the **payments** table as an example.

```python
# Check for duplicates query
query = """
        SELECT customerNumber, checkNumber, paymentDate, amount, COUNT(*) AS number_of_occurences
          FROM payments
         GROUP BY customerNumber, checkNumber
         HAVING number_of_occurences > 1;
        """
pd.read_sql(query, conn)
```

The result from the query above did not return any records, indicating that there are no duplicates in the payments table. While we can perform the same check for all tables to identify duplicates, we are confident that our database design and structure prevent duplicates from being allowed in any of the tables

### 3.1 Handle missing values
Suppose we encounter missing values in the payment table. While one option is to filter them out, this approach could result in the loss of valuable data points that might be crucial for our analysis. A more effective solution is to replace the NULL values with a placeholder, such as a string or another value determined by the analyst. In this case, we will demonstrate how to replace NULL values with 'unknown'. **NOTE**: We are not updating the database, we are reading data from the database.

```python
# Assuming missing values are in "amount" column
query = """
        SELECT customerNumber AS customer_number,
               checkNumber AS check_number,
               paymentDate AS payment_date,
               REPLACE(amount, NULL, 'unknown') AS amount
          FROM Payments;
        """
```

### 4.1 Sales Performance Analysis

We will begin by calculating the **Total Revenue** and **Total Gross Profit**, followed by an analysis of the **Monthly Sales Trend** and **Yearly Sales Trend**.
#### Total Revenue

```python
query = """
        SELECT SUM(quantityOrdered * priceEach) AS "Total Revenue"
          FROM orderdetails;
        """
pd.read_sql(query, conn)
```

#### Total Gross Profits

```python
query = """
        SELECT SUM((od.quantityOrdered * od.priceEach) -(p.buyPrice* od.quantityOrdered)) AS "Total Gross Profit"
          FROM products AS p
          JOIN orderdetails AS od
            ON p.productCode = od.productCode;
        """
pd.read_sql(query, conn)
```

Totol Gross Profit = **USD 3, 825, 880.25**
#### Monthly Trend Analysis
This analysis will help us determine seasonality, allowing us to identify the months of the year when we can expect increased demand for cars. 

```python
query = """
        SELECT strftime('%m', o.orderDate) AS date,
               COUNT(*) AS number_of_orders,
               SUM(od.quantityOrdered * od.priceEach) AS revenue
          FROM orders AS o
          JOIN orderDetails AS od
            ON o.orderNumber = od.orderNumber
         GROUP BY date
         ORDER BY date;
         """
monthly_sales_trend = pd.read_sql(query, conn)
# Rename date column to contain the names of months: "January" for "01"
mapping = {'01': 'January', '02': 'February', '03': 'March', '04': 'April', '05': 'May', '06': 'June',\
           '07': 'July', '08': 'August', '09': 'September', '10': 'October', '11': 'November', '12': 'December'}
monthly_sales_trend['date'] = monthly_sales_trend['date'].replace(mapping)
monthly_sales_trend
```

The table above indicates that the number of orders and sales revenues are highest during the months of October and November. This insight can help management ensure that sufficient stock is available during these months to meet the increased demand for cars.
#### Yearly Trend Analysis
In this section, we will analyze the yearly sales trend by calculating the **percentage change** in revenue over a three-year period: **2003**, **2004**, and **2005**.

```python
query = """
        WITH yearly_sales AS(
        SELECT strftime('%Y', o.orderDate) AS date,
               COUNT(*) AS number_of_orders,
               SUM(od.quantityOrdered * od.priceEach) AS revenue
          FROM orders AS o
          JOIN orderDetails AS od
            ON o.orderNumber = od.orderNumber
         GROUP BY date
         ORDER BY date
         )
         SELECT date, number_of_orders, revenue,
                LAG(revenue, 1, 0.0) OVER(ORDER BY date) AS previous_year_sales,
                ROUND(CASE
                    WHEN LAG(revenue) OVER(ORDER BY date) = 0 THEN NULL
                    ELSE ((revenue - LAG(revenue, 1, 0.0) OVER(ORDER BY date)) / LAG(revenue, 1, 0.0) OVER(ORDER BY date)) * 100.0 
                END, 2) AS percentage_change
           FROM yearly_sales;
         """
yearly_sales_trend = pd.read_sql(query, conn)
yearly_sales_trend
```

As shown in the table above, revenue increased by **36%** from **2003** to **2004** but decreased by **61%** in **2005**. These trends align with the changes in the number of **orders**, as **2004** had significantly more orders compared to both **2003** and **2005**.
### 4.2 Product Performance Analysis
In this section, we will address our first question: **Question 1** - Which products should we order more of or less of? This question focuses on inventory reports, including **low stock levels** and **product performance**. By answering this, we can optimize supply and enhance the user experience by ensuring that the best-selling products do not go out of stock.

To achieve this, we will first calculate the low stock levels (**demand pressure**) for each product to identify which items are understocked. Next, we will evaluate product performance to determine the top ten best-performing products. The products that are both high-performing and low in stock are the ones we should prioritize ordering more of.

Let’s proceed with the analysis by identifying the top ten products with **low stock levels**.
#### Products with low stock levels
To calculate low stock we will use a formula for calculating **demand pressure** which  in the context of inventory management, refers to the extent to which customer demand is straining or exceeding the available supply of a product. It essentially measures how "hot" or in-demand a product is.

$demand pressure = SUM(quantityOrdered)/SUM(quantityInStock)$

* **High *demand pressure* (e.g., > 1)**: This indicates that the quantity ordered significantly exceeds the quantity in stock. It suggests strong demand for your products. While this can be a positive sign, it also means you might be facing potential stockouts if you don't replenish your inventory quickly.
* ***demand pressure* close to 1**: This implies that the quantity ordered is roughly equal to the quantity in stock. It suggests a relatively balanced situation where your inventory is meeting current demand.
* **Low *demand pressure* (e.g., < 1)**: This means that the quantity ordered is less than the quantity in stock. It could indicate that demand is lower than anticipated, and you might have excess inventory.

Let us calculate demand for each product and display the top ten products with high demand pressure (**low stock**).

```python
query = """
        SELECT p.productCode AS product_code,
               p.productName AS product_name,
               p.quantityInStock AS quantity_in_stock,
               ROUND(CAST(SUM(od.quantityOrdered) AS FLOAT) / p.quantityInStock, 2) AS demand_pressure
          FROM products AS p
          JOIN orderdetails AS od
            ON p.productCode = od.productCode
         GROUP BY p.productCode
         ORDER BY demand_pressure DESC
         LIMIT 10;
        """
low_stock_products = pd.read_sql(query, conn)
low_stock_products
```

he table above highlights the top ten products with the highest demand pressure, suggesting potential understocking.  Our next step is to determine which of these products should be prioritized for restocking.  To do this, we'll identify the top ten revenue-generating products that also experience high demand pressure.
#### Product Performance by Total Revenue

```python
query = """
        SELECT p.productName AS product_name,
               COUNT(*) AS num_of_orders,
               SUM(od.quantityOrdered) AS quantity_ordered,
               SUM(od.quantityOrdered * od.priceEach) AS total_revenue
          FROM products AS p
          JOIN orderdetails AS od
            ON p.productCode = od.productCode
         GROUP BY p.productCode
         ORDER BY total_revenue DESC
         LIMIT 10;
        """
top_10_performing_products = pd.read_sql(query, conn)
top_10_performing_products
```

The top ten performing products are shown in the table above.  To determine which products require increased ordering, we will focus on those that are both understocked and high-performing.

```python
query = """
        WITH d_pressure AS(
        SELECT p.productName AS product_name,
               p.quantityInStock AS quantity_in_stock,
               ROUND(SUM(od.quantityOrdered) *1.0 / p.quantityInStock, 2) AS demand_pressure
          FROM products AS p
          JOIN orderdetails AS od
            ON p.productCode = od.productCode
         GROUP BY p.productCode
         ORDER BY demand_pressure DESC
         LIMIT 10
        ),
        p_performance AS(
        SELECT p.productName AS product_name,
               COUNT(*) AS num_of_orders,
               SUM(od.quantityOrdered) AS quantity_ordered,
               SUM(od.quantityOrdered * od.priceEach) AS total_revenue
          FROM products AS p
          JOIN orderdetails AS od
            ON p.productCode = od.productCode
         GROUP BY p.productCode
         ORDER BY total_revenue DESC
         LIMIT 10
        )
        SELECT product_name
          FROM d_pressure
         WHERE product_name IN (SELECT product_name
                                  FROM p_performance);
        """
pd.read_sql(query, conn)
```

The product that needs restocking is the **1968 Ford Mustang** as this product is both a high performer by revenue and has high **demand pressure** meaning it is mostly ordered by customers.
### 4.3 Customer Insights
#### Top performing customers

```python
query = """
        SELECT c.customerNumber AS customer_id,
               c. customerName AS customer_name,
               p.amount AS payment_amount
          FROM customers AS c
          JOIN payments AS p
            ON c.customerNumber = p.customerNumber
         GROUP BY customer_id
         ORDER BY payment_amount DESC
         LIMIT 5;
        """
top_5_performing_customers = pd.read_sql(query, conn)
top_5_performing_customers
```

#### Customer Segmentation
Customer segmentation is the process of dividing a company's customers into groups based on common characteristics, behaviors, or needs. This allows businesses to tailor their marketing strategies, products, and services to better meet the needs of each segment, ultimately improving customer satisfaction and profitability. In our case, we will segment customers based on their country of origin and determine the top five countries that contribute the most to revenue generation of our company.

```python
query = """
        SELECT c.country,
               SUM(od.quantityOrdered) AS total_products_sold,
               SUM(od.quantityOrdered * od.priceEach) AS total_revenue
          FROM customers AS c
          JOIN orders AS o
            ON c.customerNumber = o.customerNumber
          JOIN orderdetails AS od
            ON o.orderNumber = od.orderNumber
         GROUP BY country
         ORDER BY total_revenue DESC
         LIMIT 5;
        """
top_5_countries = pd.read_sql(query, conn)
top_5_countries
```

```python
query = """
        SELECT o.customerNumber AS customer_number,
               ROUND(SUM((od.quantityOrdered*od.priceEach)-(od.quantityOrdered*p.buyPrice)),2) AS profit
          FROM orders AS o
          JOIN orderdetails AS od
            ON o.orderNumber = od.orderNumber
          JOIN products AS p
            ON p.productCode = od.productCode
         GROUP BY customer_number;
        """
profits_by_customer_number = pd.read_sql(query, conn)
profits_by_customer_number.head()
```

```python
query = """
        WITH customer_profit AS(
      SELECT o.customerNumber AS customer_number,
             ROUND(SUM((od.quantityOrdered*od.priceEach)-(od.quantityOrdered*p.buyPrice)),2) AS profit
        FROM orders AS o
        JOIN orderdetails AS od
          ON o.orderNumber = od.orderNumber
        JOIN products AS p
          ON p.productCode = od.productCode
       GROUP BY customer_number
       )
      SELECT c.contactFirstName||' '||c.contactLastName AS customer_name,
             c.city, c.country,
             cp.profit
        FROM customers AS c
        JOIN customer_profit AS cp
          ON c.customerNumber = cp.customer_number
       ORDER BY profit DESC
       LIMIT 5;
        """
VIP_customers = pd.read_sql(query, conn)
VIP_customers
```

```python
query = """
        WITH customer_profit AS(
      SELECT o.customerNumber AS customer_number,
             ROUND(SUM((od.quantityOrdered*od.priceEach)-(od.quantityOrdered*p.buyPrice)),2) AS profit
        FROM orders AS o
        JOIN orderdetails AS od
          ON o.orderNumber = od.orderNumber
        JOIN products AS p
          ON p.productCode = od.productCode
       GROUP BY customer_number
       )
      SELECT c.contactFirstName||' '||c.contactLastName AS customer_name,
             c.city, c.country,
             cp.profit
        FROM customers AS c
        JOIN customer_profit AS cp
          ON c.customerNumber = cp.customer_number
       ORDER BY profit
       LIMIT 5;
        """
least_engaged_customers = pd.read_sql(query, conn)
least_engaged_customers
```

### 4.4 Employee Performance Analysis
#### Employee sales contributions

```python
query = """
        SELECT e.firstName||' '||e.lastName AS sales_rep_name,
               SUM(od.quantityOrdered) AS total_cars_sold,
               SUM(od.quantityOrdered * od.priceEach) AS total_revenue_generated
          FROM employees AS e
          JOIN customers AS c
            ON e.employeeNumber = c.salesRepEmployeeNumber
          JOIN orders AS o
            ON c.customerNumber = o.customerNumber
          JOIN orderdetails AS od
            ON o.orderNumber = od.orderNumber
         GROUP BY sales_rep_name
         ORDER BY total_revenue_generated DESC
         LIMIT 10;
        """
top_ten_sales_rep = pd.read_sql(query, conn)
top_ten_sales_rep
```
