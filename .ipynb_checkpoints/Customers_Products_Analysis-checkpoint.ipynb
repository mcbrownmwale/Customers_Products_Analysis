# Sales Performance Analysis for a Retail Business

## 1. Introduction
This project will focus on analyzing sales performance using SQL. It will cover key metrics such as revenue trends, customer spending behavior, product performance, and employee sales contributions.
### 1.1 Business Problem
A retail company wants to improve its sales performance by understanding key trends in customer purchasing behavior, product demand, and sales team efficiency. However, they lack clear insights into which products are the best sellers, which customers contribute the most revenue, and how sales performance varies over time. Without this data-driven approach, the company may miss opportunities to optimize pricing, inventory, and marketing strategies.
### 1.2 Project Objectives
This project aims to:

1. **Analyze sales trends** over time to identify peak sales periods.
2. **Evaluate customer spending behavio**r to identify high-value customers.
3. **Determine the best-selling products** to optimize inventory management.
4. **Assess employee sales performance** to reward high performers and identify areas for improvement.
5. **Analyze payment trends** to ensure financial stability and reduce delayed payments.
### 1.3 Identify Key Performance Indicators (KPIs)
    
To measure success, the project will track the following KPIs:

**Sales Performance**
* Total Revenue: Sum of all sales transactions.
* Monthly Revenue Growth: Percentage change in sales month-over-month.
* Total Orders: Number of orders placed over a given period.
    
**Customer Insights**
* Top 5 Customers by Revenue: Identifying customers with the highest lifetime value.
* Average Order Value (AOV): Total revenue divided by the number of orders.
    
**Product Performance**
* Best-Selling Products: Products generating the most revenue.
* Units Sold per Product: Number of units sold per product category.
    
**Employee Performance**
* Sales per Employee: Total revenue generated by each sales representative.
* Orders Handled per Employee: Number of orders managed by each employee.

**Financial Analysis**
* Total Payments Received: Sum of all payments made by customers.
* Average Payment Amount: Total payments divided by the number of transactions.
* Delayed Payments: Orders with a payment received after the due date.

## 4. Data Analysis
### 4.1 Sales Trend Analysis

```python
query = """
        SELECT strftime('%Y-%m', o.orderDate) AS date,
               COUNT(*) AS number_of_orders,
               SUM(od.quantityOrdered * od.priceEach) AS revenue
          FROM orders AS o
          JOIN orderDetails AS od
            ON o.orderNumber = od.orderNumber
         GROUP BY date
         ORDER BY date;
         """
sales_trend = pd.read_sql(query, conn)
sales_trend
```

```python
sales_trend.plot()
plt.show()
```

### 4.2 Top-performing products

```python
query = """
        SELECT p.productName AS product_name,
               COUNT(*) AS num_of_orders,
               SUM(od.quantityOrdered) AS quantity_ordered,
               SUM(od.quantityOrdered * od.priceEach) AS total_revenue
          FROM products AS p
          JOIN orderdetails AS od
            ON p.productCode = od.productCode
         GROUP BY product_name
         ORDER BY num_of_orders DESC, total_revenue DESC
         LIMIT 5;
        """
top_10_performing_products = pd.read_sql(query, conn)
top_10_performing_products
```

### 4.3 Top performing customers

```python
query = """
        SELECT c.customerNumber AS customer_id,
               c. customerName AS customer_name,
               p.amount AS payment_amount
          FROM customers AS c
          JOIN payments AS p
            ON c.customerNumber = p.customerNumber
         GROUP BY customer_id
         ORDER BY payment_amount DESC
         LIMIT 5;
        """
top_5_performing_customers = pd.read_sql(query, conn)
top_5_performing_customers
```

### 4.3 Customer Segmentation
Customer segmentation is the process of dividing a company's customers into groups based on common characteristics, behaviors, or needs. This allows businesses to tailor their marketing strategies, products, and services to better meet the needs of each segment, ultimately improving customer satisfaction and profitability. In our case, we will segment customers based on their country of origin and determine the top five countries that contribute the most to revenue generation of our company.

```python
query = """
        SELECT c.country,
               SUM(od.quantityOrdered) AS total_products_sold,
               SUM(od.quantityOrdered * od.priceEach) AS total_revenue
          FROM customers AS c
          JOIN orders AS o
            ON c.customerNumber = o.customerNumber
          JOIN orderdetails AS od
            ON o.orderNumber = od.orderNumber
         GROUP BY country
         ORDER BY total_revenue DESC
         LIMIT 5;
        """
top_5_countries = pd.read_sql(query, conn)
top_5_countries
```

### 4.4 Employee sales contributions

```python
query = """
        SELECT e.firstName||' '||e.lastName AS sales_rep_name,
               SUM(od.quantityOrdered) AS total_cars_sold,
               SUM(od.quantityOrdered * od.priceEach) AS total_revenue_generated
          FROM employees AS e
          JOIN customers AS c
            ON e.employeeNumber = c.salesRepEmployeeNumber
          JOIN orders AS o
            ON c.customerNumber = o.customerNumber
          JOIN orderdetails AS od
            ON o.orderNumber = od.orderNumber
         GROUP BY sales_rep_name
         ORDER BY total_revenue_generated DESC
         LIMIT 10;
        """
top_ten_sales_rep = pd.read_sql(query, conn)
top_ten_sales_rep
```
